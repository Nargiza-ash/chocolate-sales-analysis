```python
import pandas as pd
df=pd.read_csv("Chocolate Sales (2).csv")
df.head(10)
```

```python
df.groupby('Sales Person').size()
```

```python
df['Date']=pd.to_datetime(df['Date'], format='%d/%m/%Y', errors='coerce')
```

```python
df.info()
```

```python
df['Amount']=(
    df['Amount']
    .astype(str)
    .str.replace('$', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.strip()
    .astype(float)
)
```

```python
sales_by_person = (
    df.groupby('Sales Person')
    .agg(
        Количество_продаж=('Amount', 'count'),
        Сумма_продаж=('Amount', 'sum'),
        Коробок_отгружено=('Boxes Shipped', 'sum')
    )
    .reset_index()
)
sales_by_person 
```

```python
pivot_sales = pd.pivot_table(
    df,
    values='Amount',
    index='Sales Person',
    columns='Country',
    aggfunc='sum'
)
pivot_sales
```

```python
pivot_avg = pd.pivot_table(
    df,
    values='Amount',
    index='Sales Person',
    columns='Country',
    aggfunc='mean'
)
pivot_avg 
```

```python
df.isna().sum()
#выполнена проверка наличия пропущенных значений.
```

```python
df['Amount'] = df['Amount'].fillna(0)
#пропущенные значения в сумме продаж заменены на 0.
```

```python
df['Sales Person'] = df['Sales Person'].str.upper()
#ФИО продавцов приведены к единому формату.
```

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df['Month'] = df['Date'].dt.month
#столбец даты преобразован в формат datetime и выделен месяц.
```

```python
df.info()
```

```python
df.head()
```

```python
df_2022 = df[df['Date'].dt.year == 2022]
df_2023 = df[df['Date'].dt.year == 2023]

df_all = pd.concat([df_2022, df_2023], ignore_index=True)
#данные за разные периоды объединены в один датафрейм с помощью concat.
df_all
```

```python
top_3_chocolate = (
    df
    .groupby('Product')
    .agg(Доход=('Amount', 'sum'))
    .sort_values('Доход', ascending=False)
    .head(3)
    .reset_index()
)
top_3_chocolate
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year

pivot_year = pd.pivot_table(
    df,
    values='Amount',
    index='Year',
    aggfunc='sum'
)
pivot_year
```
